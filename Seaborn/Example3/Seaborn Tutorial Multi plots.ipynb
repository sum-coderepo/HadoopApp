# Controlling Ploted Figure Aesthetics

```python
def simplot(flip = 1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x, np.sin(x+i*0.5)* (7-i)*flip)
        
```

```python
simplot(-1)
```

```python
sns.set_style('ticks',{'axes.grid' : True,'xtick.direction': 'in'}) # For bottom ticks
simplot(1)
sns.despine() # tp remove the left and Right border
```
